```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define tickers and timeframe
  # SPY = S&P 500 ETF
  # RSP = S&P 500 Equal Weighted ETF
  # set today as end date
  # set 1 year from today as start date
tickers = ["SPY", "RSP"]
end_date = pd.Timestamp.now()
start_date = end_date - pd.DateOffset(years=1)

# Fetch adjusted closing prices
  # yfinance defaults auto_adjust=True that define closeing prices as 'Close'
data = yf.download(tickers, start=start_date, end=end_date, auto_adjust=True, progress=False)['Close']

# Calculate percentage change relative to start_date
  # subtract 1 and multiply by 100 to find % change
pct_change = (data / data.iloc[0] - 1) * 100

# Generate plot
plt.figure(figsize=(12, 6))
plt.plot(pct_change.index, pct_change['SPY'], label='SPY (S&P 500 ETF)', linewidth=2)
plt.plot(pct_change.index, pct_change['RSP'], label='RSP (S&P 500 Equal Weighted ETF)', linewidth=1.5)

plt.title('SPY vs RSP Performance Comparison (Past 1 Year)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Percentage Change (%)', fontsize=12)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8) # Reference line at 0%
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```
